```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Krishna\Desktop\Python\us_customer_data 1.csv")
```

```python
#Registration date is not in correct format
df['registration_date']=pd.to_datetime(df['registration_date'],format='%d-%m-%Y',errors='coerce')
```

```python
#standardizing contact details: Phone Number !!run only once
df['phone']=df['phone'].str.replace(r'-|[\0]|[a-zA-Z]|[\(\)]',"",regex=True)
df['phone']=df['phone'].str.slice(0,10)

```

```python
df['phone']=df['phone'].apply(lambda x: np.nan if len(str(x))<10 else x)
```

```python
df['phone']=df['phone'].fillna('9999999999')
```

```python
df['phone']='+91 '+df['phone']
```

```python
#Standardizing the contact details : Mail
df['email']=df['email'].fillna(df['name']+'@gmail.com')
df['email']=df['email'].str.lower()
df['email']=df['email'].str.replace(" ","")
```
